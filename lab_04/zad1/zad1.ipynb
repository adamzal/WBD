```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
```

```python
lda = LDA(n_components=2)
X_r2 = lda.fit(X, y).transform(X)
```

```python
print('explained variance ratio (first two components in PCA): %s' % str(pca.explained_variance_ratio_))
```

```python
print('explained variance ratio (first two components in LDA): %s' % str(lda.explained_variance_ratio_))
```

```python
X_r2.shape
```

```python
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw=2

for color, i, target_name in zip(colors, [0,1,2], target_names):
    plt.scatter(X_r[y==i,0], X_r[y==i,1], c=color, alpha=.8, lw=lw,label=target_name)
    print(X_r[y==i,0])
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title("PCA of IRIS dataset")

plt.figure()
for color, i, target_name in zip(colors, [0,1,2], target_names):
    plt.scatter(X_r2[y==i,0], X_r2[y==i,1], c=color, alpha=.8, lw=lw,label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title("LDA of IRIS dataset")
```
